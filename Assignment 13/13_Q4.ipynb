```python
import numpy as np

# a) 10 equally spaced numbers between 0 and 5
a = np.linspace(0, 5, 10)

# b) 15 equally spaced numbers between -10 and 10
b = np.linspace(-10, 10, 15)

print("0 to 5 (10 numbers):")
print(a)
print("Length:", len(a))

print("\n-10 to 10 (15 numbers):")
print(b)
print("Length:", len(b))
```
