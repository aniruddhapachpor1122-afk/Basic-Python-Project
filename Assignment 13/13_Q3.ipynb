```python
import numpy as np

# a) Numbers from 0 to 20
a = np.arange(0, 21)

# b) Even numbers from 10 to 50
b = np.arange(10, 51, 2)

# c) Numbers from 5 to 100 with step 5
c = np.arange(5, 101, 5)

print("0 to 20:")
print(a)

print("\nEven Numbers (10 to 50):")
print(b)

print("\n5 to 100 (Step 5):")
print(c)
```
