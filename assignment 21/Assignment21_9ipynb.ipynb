```python
# Import required libraries

import streamlit as st     # Web app
import pandas as pd        # Data handling
import joblib              # Load saved model
```

```python
# Load model and preprocessing files

model = joblib.load("LR_model.pkl")
scaler = joblib.load("scaler.pkl")
encoded_columns = joblib.load("columns.pkl")
```

```python
# Page configuration

st.set_page_config(
    page_title="Ford Car Price Predictor",
    layout="centered"
)
```

```python
st.title("Ford Car Price Predictor")

st.write("Enter the car details below to predict its selling price.")
```

```python
year = st.number_input("Manufacturing Year", 1990, 2025, 2018)

mileage = st.number_input("Mileage", 0, 300000, 50000)

tax = st.number_input("Road Tax", 0, 1000, 150)

mpg = st.number_input("MPG", 0.0, 100.0, 55.0)

engineSize = st.number_input("Engine Size", 0.8, 6.0, 1.5)
```

```python
transmission = st.selectbox(
    "Transmission",
    ["Automatic", "Manual", "Semi-Auto"]
)

fuelType = st.selectbox(
    "Fuel Type",
    ["Petrol", "Diesel", "Hybrid", "Electric"]
)
```

```python
model_name = st.text_input("Car Model")

predict = st.button("Predict Price")
```

```python
if predict:

    data = pd.DataFrame({
        "model":[model_name],
        "year":[year],
        "transmission":[transmission],
        "mileage":[mileage],
        "fuelType":[fuelType],
        "tax":[tax],
        "mpg":[mpg],
        "engineSize":[engineSize]
    })

    data = pd.get_dummies(data)

    data = data.reindex(columns=encoded_columns, fill_value=0)
```
