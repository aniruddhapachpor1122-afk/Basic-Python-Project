```python
import numpy as np
import pandas as pd
from google.colab import files
uploaded=files.upload()
```

```python
filepath='/content/Colth.csv'
df=pd.read_csv(filepath)
df
```

```python
df.head()
```

```python
X = df.drop('Price', axis=1)
y = df['Price']
print("\nShape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
cat_cols = X.select_dtypes(include=['object']).columns
cat_cols
X_encoded = pd.get_dummies(X, columns=cat_cols).astype(int)
X_encoded.head()
```

```python
from sklearn.preprocessing import StandardScaler
num_cols = X_encoded.select_dtypes(include=np.number).columns
scaler = StandardScaler()
X_encoded[num_cols] = scaler.fit_transform(X_encoded[num_cols])
X_encoded.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.33, random_state=42)
print("\nShapes:")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
print("\nIntercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

```python
y_pred = model.predict(X_test)
print("\nFirst 10 Predictions:", y_pred[:10])
print("First 10 Actual:", y_test.values[:10])
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print("\nR2 Score:", r2)
if r2 > 0.8:
    print("Model is performing very well.")
elif r2 > 0.6:
    print("Model performance is good.")
else:
    print("Model needs improvement.")
```

```python
import joblib
joblib.dump(model, 'LR_cloth.pkl')
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(X_encoded.columns.tolist(), 'columns.pkl')
print("\nModel and files saved successfully!")
```
