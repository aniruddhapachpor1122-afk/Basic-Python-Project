```python
import numpy as np

arr = np.arange(1, 25)

print("Original Array:\n", arr)
print("Shape:", arr.shape)

a = arr.reshape(4, 6)
print("\nReshape (4,6):")
print(a)
print("Shape:", a.shape)

b = arr.reshape(3, 8)
print("\nReshape (3,8):")
print(b)
print("Shape:", b.shape)

c = arr.reshape(2, 3, 4)
print("\nReshape (2,3,4):")
print(c)
print("Shape:", c.shape)
```
