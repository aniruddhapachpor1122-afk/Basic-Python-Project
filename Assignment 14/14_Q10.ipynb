```python
import numpy as np

# Generate marks for 10 students (5 subjects)
marks = np.random.randint(30, 101, (10, 5))

print("Student Marks:\n")
print(marks)

# Total and Average of each student
total = marks.sum(axis=1)
average = marks.mean(axis=1)

print("\nTotal Marks:")
print(total)

print("\nAverage Marks:")
print(average)

# Highest and Lowest scorer
highest = np.argmax(total)
lowest = np.argmin(total)

print("\nHighest Scoring Student Index:", highest)
print("Total Marks:", total[highest])

print("\nLowest Scoring Student Index:", lowest)
print("Total Marks:", total[lowest])

# Overall class statistics
print("\nClass Mean:", marks.mean())
print("Class Standard Deviation:", marks.std())

# Top 3 students
top3 = np.argsort(total)[-3:]

print("\nTop 3 Students Marks:")
print(marks[top3])

# Comments:
# sum(axis=1) -> Total marks of each student.
# mean(axis=1) -> Average marks of each student.
# argmax() -> Index of highest scorer.
# argmin() -> Index of lowest scorer.
# argsort()[-3:] -> Indices of top 3 students.
```
