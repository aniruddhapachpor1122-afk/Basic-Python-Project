```python
import numpy as np

arr = np.random.randn(6, 6)

print("Matrix:\n", arr)
print("Shape:", arr.shape)
print("Size:", arr.size)
print("Data Type:", arr.dtype)

print("Index of Maximum:", np.argmax(arr))
print("Index of Minimum:", np.argmin(arr))

print("\nTop-left 3x3 Submatrix:\n", arr[:3, :3])

arr = np.abs(arr)

print("\nModified Matrix:\n", arr)
print("Mean:", arr.mean())
```
